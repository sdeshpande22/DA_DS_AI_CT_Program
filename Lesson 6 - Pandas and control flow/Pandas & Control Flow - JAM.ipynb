```python
list_of_offers = [5,6,8,10,11,2,5.5]
```

```python
for o in list_of_offers:
    if o >= 8:
        print("Accept the offer")
    else:
        print("Reject the Offer, Call to Them and Neg")
```

```python
def kanth_offers(data):
    for o in data:
        if o >= 8:
            print("Accept the offer")
        else:
            print("Reject the Offer, Call to Them and Neg")
```

```python
kanth_offers([4,3])
```

```python
kanth_offers([5,8])
```

```python
def kanth_bank(cibil_score):
    for o in cibil_score:
        if o >= 700:
            print("Approve the loan")
        else:
            print("Reject the loan")
```

```python
kanth_bank([560, 760,800,340])
```
